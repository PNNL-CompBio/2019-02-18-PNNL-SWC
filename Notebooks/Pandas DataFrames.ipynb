# Use `DataFrame.iloc[..., ...]` to selet values by their (entry) position

```python
import pandas as pd
data = pd.read_csv('data/gapminder_gdp_europe.csv', 
                  index_col='country'
                  )
data
```

```python
data.columns
```

```python
data.index
```

```python
data.iloc[0,0]
```

```python
data.T.loc[:,  0:3]
```

```python
data.iloc[:, 0:3]
```

```python
data.loc['Albania','gdpPercap_1952']
```

```python
#display(data.loc['Albania', :].to_frame('Albania')), 
data.loc[:,['gdpPercap_1952','gdpPercap_1957']]
```

```python
data.loc[:,'gdpPercap_1952':'gdpPercap_1962']
```

# `iloc` has different semantics than `loc` when it comes to slicing

```python
data.iloc[:,0:3]
```

```python
data.loc['Albania'].to_frame('Albania')
```

```python
data.loc[:, 'gdpPercap_1952'].to_frame('gdpPercap_1952')
```

```python
data['gdpPercap_1952'].to_frame('gdpPercap_1952')
```
